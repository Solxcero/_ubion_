```python
import pandas as pd
```

```python
sales_df = pd.read_csv("./Sales Records.csv")
sales_df
```

```python
# 컬럼 삭제(Ship Date, Order ID, Sales Channel,Total Revenue ) --> 원본데이터 유지(새로운 변수)
# 컬러명 수정 -->  원본데이터 유지 x (지역,국가,상품타입,배송순위,주문날짜,판매개수,단가,원가)
# Country 기준으로 sort
# index 값 초기화 
```

```python
sales_df_1 = sales_df.drop(["Ship Date","Order ID","Sales Channel","Total Revenue"],axis=1)
```

```python
sales_df_1.columns=["지역","국가","상품타입","배송순위","주문날짜","판매개수","단가","원가","총비용","총이윤"]
```

```python
sales_df_1.sort_values("국가",inplace=True)
```

```python
sales_df_1.reset_index(drop=True,inplace=True)  #drop:기존의 index 삭제 clccdasdfcccccasdffdasdf
sales_df_1.head(10)
```

```python
#총이윤 정수형으로 바꾸기
sales_df_1["총이윤"]=sales_df_1["총이윤"].astype(int)
```

```python
sales_df_1
```

```python
#총비용 반올림해서 정수형으로 round: 반올림(소수점자리수)
sales_df_1["총비용"]=sales_df_1["총비용"].round(0).astype(int)
sales_df_1.head()
```

```python
#원가 소수점 하나 남기고 반올림
sales_df_1["원가"]=sales_df_1["원가"].round(1).astype(int)
```

```python
#round에 음수를 입력하면?=> 정수 자리서 부터 반올림 시작 
```

```python
#주문날짜 변경
sales_df_1["주문날짜"]=pd.to_datetime(sales_df_1["주문날짜"],
                                      format="%m/%d/%Y")  #대소문자 상관있음
sales_df_1.head()                                     
```

```python
#group by
sales_df_1.groupby("국가").sum()["총이윤"]
```

```python
sales_df_1.groupby(["지역","국가","상품타입"]).sum()["총이윤"]  #sum 대신 mean 써도 됨 
```

```python
#하나만 뽑기  #get_group 뒤에는 튜플 형식으로 입력 
sales_df_1.groupby(["국가","상품타입"]).get_group(("South Korea","Clothes"))
```

```python
#get_group 에 여러 국가 집어넣는 방법 
sales_df_2 = sales_df_1[sales_df_1["국가"].isin(["South Korea","China","Togo"])]
sales_df_2.groupby("국가").mean()["총이윤"]
```
