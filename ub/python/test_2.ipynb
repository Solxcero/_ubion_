```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4]) #리스트 값은 y값으로 . x축은 자동으로 0,1,2,3,4 채워짐 
plt.show()
```

```python
plt.plot([1,2,3,4,],[1,4,5,10])
plt.show()   # 첫 리스트의 값:x축 , 두번째 리스트 값:y축
```

```python
# 옵션 종류들
# b:blue/g:green/r:red/c:cyan/m:magenta/y:yellow/k:black/w:white => 헥스코드로 색상입력해도 됨
# 선 종류
plt.plot([1,2,3,4],[1,4,5,10],'c--')
```

```python
plt.plot([1,2,3,4],[1,4,5,10],'--r')
plt.plot([1,2,3,4],[5,3,7,1],'ob')
plt.plot([1,2,3,4],[8,2,5,3],'.m')
```

```python
# x,y 축 레이블 추가 => plt.xlabel(), plt.ylabel()
# 그래프 제목 추가 => plt.title()
```

```python
#범례 추가 
# plt.legend()

plt.plot(["Maths","Science","History"],[90,95,60],'r-.')
plt.plot(["Maths","Science","History"],[65,70,95],'c--')
plt.xlabel("Sub")
plt.ylabel("Score")
plt.title("lol")
plt.legend(["A","B"], loc = 5) #loc="" 범례 위치 지정 => string 이나 code값으로 지정가능  
plt.show()
```

```python
#서브플롯-여러 그래프 한번에 
#plt.subplot(nrow,ncol,pos)  : nrow:행의 수 , ncol:열의 수 , pos: 위치

plt.subplot(2,1,1)  #확인
plt.plot([3,5,1,7])
plt.subplot(2,1,2)  #확인 
plt.plot([1,2,3,4,])
plt.show()
```

```python
x = [1,2,3]
y = [10,20,30]

plt.bar(x,y,color=['c','y'],width=0.4)  #각 바 마다 색상 지정할 때는 리스트로 
plt.xticks(x,["Maths","Science","History"])  #xticks() 이용하여 x축의 레이블 값 변경 
plt.show()
```

```python
#산점도 

x = np.random.rand(50)
y = np.random.rand(50)

area = (20*np.random.rand(50))**2.2   #산점도 default값
colors = np.random.rand(50)

plt.scatter(x,y, s=area, c=colors)
plt.show()
```
