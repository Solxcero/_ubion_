### 3번 문제
> 3개의 숫자를 입력하면 (input_a * input_b + input_c)와 (input_a + input_b * input_c)의 차이를 나타내는 함수를 만드시오.
1. 함수명은 diff_sum()이라 지정
2. 함수에 매개변수를 3개 지정(input_a, input_b, input_c)
3. 위 조건에서 큰수에서 작은 수를 빼는 if문 작성 
4. 3개의 숫자는 input를 사용하여 입력 값을 받을 것

```python
def diff_sum(input_a, input_b, input_c):
    
    x = input_a * input_b + input_c
    y = input_a + input_b * input_c
    if x > y:
        return x - y
    elif x < y:
        return y - x
    else:
        return "input again"
diff_sum(int(input("a:")),int(input("b:")),int(input("c:"))) 

```

### 5번 문제
> csv파일을 읽어온 뒤 데이터프레임을 수정하는 Class를 만들어보자.

1. class를 생성할 때, 매개 변수를 이용하여 csv 파일 주소를 받아 온 후 csv파일 읽기(파일 주소는 상대경로로 지정)
2. class 내의 함수 _sort()를 생성 후 컬럼의 이름이나 컬럼의 위치(예 : n번째 컬럼)를 받아오는 매개 변수를 생성. 그 값을 기준으로 sort를 하는 함수 생성(매개 변수가 위치(int)인지 이름(str)인지 체크하여 함수 구성, 오름차순 정렬)
3. class 내의 함수 _drop()을 생성하여 매개 변수를 이용하여 컬럼의 이름을 받아온 후 해당 컬럼을 삭제하는 함수 생성
4. class 내의 함수 _reset()을 생성하여 인덱스의 값을 초기화 하고 기존의 인덱스 값은 삭제하는 함수 생성
5. 모든 과정에서 원본의 데이터를 수정

```python
import pandas as pd
```

```python
class Class_1():
    def __init__(self, input_url):
        self.url = input_url

    def _sort(self, input_sort,input_type):
        self.dt = pd.read_csv(self.url)
        if input_type == "s":  #str
            self.dt.sort_values(input_sort, inplace = True)
            return self.dt
        elif input_type == "i":  #int
            self.dt.sort_values(self.dt.columns[(input_sort)], inplace = True)
            return self.dt


    def _drop(self,input_drop=[]):
        self.dt.drop(input_drop,axis=1,inplace=True)
        return self.dt

    def _reset(self):
        self.dt.reset_index(drop=True, inplace=True)
        return self.dt
        
```

```python
dt = Class_1("../csv/Sales Records.csv")
```

```python
dt._sort("Order Date",'s')
```

```python
dt._sort(0,'i')
```

```python
dt._drop("Sales Channel")
```

```python
dt._reset()
```
