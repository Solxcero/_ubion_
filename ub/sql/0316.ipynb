```python
# 읽는 순서 SELECT 다음이 FROM  그 다음이 컬럼작업 
sql_17='''
SELECT
emp.ENAME AS 사원명,
CASE    
    WHEN emp.JOB = "ANALYST" THEN "분석가"
    WHEN emp.JOB = "CLERK" THEN "점원"
    WHEN emp.JOB = "MANAGER" THEN "관리자"
    WHEN emp.JOB = "SALESMAN" THEN "판매사원"
    WHEN emp.JOB = "PRESIDENT" THEN "사장"
    END 
    AS 업무,
IF (emp.SAL >= 2450, 'HIGH', 'LOW') AS 월급,
CONCAT(emp.DEPTNO,'-',dept.DNAME) AS "부서번호-부서명",
dept.LOC AS 위치
FROM emp LEFT JOIN dept
ON emp.DEPTNO = dept.DEPTNO   
ORDER BY `사원명` ASC   
'''
result_17=_dt.executeAll(sql_17)
df_17 = pd.DataFrame(result_17)
df_17
```

### emp table
> 사원 정보를 담고 있는 테이블

### dept table
> 부서의 정보를 담고 있는 테이블 

### 부서의 이름이 'sales'인 소속 사원들의 리스트 출력

```python
sql_20 = '''
            SELECT 
            ENAME 
            FROM emp
            LEFT JOIN dept
            ON emp.DEPTNO=dept.DEPTNO
            WHERE dept.DNAME = "sales"
            ORDER BY ENAME ASC '''
result_20 = _dt.executeAll(sql_20)
pd.DataFrame(result_20)
```

```python
sql_21 = '''
            SELECT 
            ENAME 
            FROM emp
            WHERE DEPTNO = 
            (
                SELECT
                DEPTNO FROM dept
                WHERE DNAME='sales'
            )  
            ORDER BY ENAME ASC
            '''
result_21 = _dt.executeAll(sql_21)
pd.DataFrame(result_21)
```

### 부서이름이 'sales'가 아닌 소속 사원들의 리스트 출력

```python
sql_22 = '''
            SELECT 
            ENAME 
            FROM emp
            WHERE DEPTNO != 
            (
                SELECT
                DEPTNO FROM dept
                WHERE DNAME='sales'
            )  
            ORDER BY ENAME ASC
            '''
result_22 = _dt.executeAll(sql_22)
pd.DataFrame(result_22)
```

### 부서의 위치가 'NEW YORK', 'DALLAS'가 아닌 소속 사원들의 리스트 . != 쓰지마 

```python
sql_23 = '''
            SELECT 
            ENAME 
            FROM emp
            WHERE DEPTNO IN
            (
                SELECT
                DEPTNO FROM dept
                WHERE LOC NOT IN('NEW YORK','DALLAS')
                
            )            
            ORDER BY ENAME ASC
            '''
            # 1242 ERROR 뜨는 이유 : 내부 select문에서 출력되는 값이 여러개라 '=' 쓰면 인식을 못해..
            # DEPTNO = any 써도 됨. 
result_23 = _dt.executeAll(sql_23)
pd.DataFrame(result_23)
```

### dname=operation 이고 loc=newyork 

```python
sql_24 = '''
            SELECT 
            ENAME 
            FROM emp
            WHERE DEPTNO =
            (
                SELECT
                DEPTNO FROM dept
                WHERE DNAME='OPERATIONS'
            )
            OR DEPTNO =
            (
                SELECT
                DEPTNO FROM dept
                WHERE LOC='NEW YORK'
            )
            ORDER BY ENAME ASC
            '''
result_24 = _dt.executeAll(sql_24)
pd.DataFrame(result_24)
```
