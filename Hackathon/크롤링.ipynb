```python
# 금융공학레시피 참고하여 짠거
from urllib.request import urlopen
import bs4
import datetime as dt
import numpy as np

source = urlopen('https://finance.naver.com/sise/sise_group_detail.naver?type=upjong&no=281').read()  # 281 대신 다른 업종 페이지 쓰면 됨
source = bs4.BeautifulSoup(source,from_encoding='euc-kr')
print(source.prettify())
```

```python
string = source.find_all('div',class_='name_area')
string
```

```python
name = []  #종목명만 뽑아내기 
for link in string:
    name.append(link.get_text())
    
name
```

```python
name_1 = []  # *(코스닥상장) 표시 지우기 
for i in range(len(name)):
    name_1.append(name[i].strip(' *'))
    
name_1
```

```python
# 멘토님이 주신 코드 

import numpy as np 
import requests
from bs4 import BeautifulSoup
import time
# 업종 261~ 339 까지 존재
rqg_no = np.arange(261,340)
rqg_no = [str(x) for x in rqg_no]
for number1 in rqg_no:
    url = 'https://finance.naver.com/sise/sise_group_detail.naver?type=upjong&no='+number1
    res = requests.get(url)
    soup = BeautifulSoup(res.content,'html.parser')
    section = soup.find('tbody')
    items= section.find_all('tr', onmouseover="mouseOver(this)")
    for item in items :
        basic_info = item.get_text()
        sinfo = basic_info.split("\n")
        print(sinfo)
        print(sinfo[1],sinfo[2],sinfo[5])
    time.sleep(1)
    break; # 요거 빼시고 위에서 원하는 것들을 정해서 데이터 프레임에 저장 -> merge? 하시면 될꺼 같습니다. (요건 멘토님이 써주신거 ^0^)
    
```
