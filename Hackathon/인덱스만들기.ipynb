```python
from pykrx import stock
import time
```

```python
import pandas as pd
```

```python
hc = ["137310","214150","085370","145720","290650"]  #헬스케어
bio = ["145720","068270","302440","091990","000100"]   #제약
tr = ["035250","008770","034230","032350","039130"] #여행
flight = ["003490","180640","020560","089590","272450"]     #항공
express = ["000120","086280","002320","004360","004140"] #운송화물
makeup= ["051900","090430","048410","002790","161890"] #화장품
```

```python
df_makeup_tt = pd.DataFrame()
for code in makeup:
    df = stock.get_market_cap("20200310", "20220228", code)['시가총액']
    
    df_makeup_tt = pd.concat([df_makeup_tt, df],axis=1)
    df_makeup_tt.rename(columns={'시가총액':f'{code}시가총액'},inplace=True)
    time.sleep(1)

df_makeup_tt['Sum'] = df_makeup_tt.sum(axis=1)
df_makeup_tt['index'] = (df_makeup_tt['Sum']/df_makeup_tt['Sum'].iloc[0]) *100
df_makeup_tt
```

```python
df_travel_tt = pd.DataFrame()
for code in tr:
    df = stock.get_market_cap("20200310", "20220228", code)['시가총액']
    
    df_travel_tt = pd.concat([df_travel_tt, df],axis=1)
    df_travel_tt.rename(columns={'시가총액':f'{code}시가총액'},inplace=True)
    time.sleep(1)

df_travel_tt['Sum'] = df_travel_tt.sum(axis=1)
df_travel_tt['index'] = (df_travel_tt['Sum']/df_travel_tt['Sum'].iloc[0]) *100
    
df_travel_tt
```

```python
df_flight_tt = pd.DataFrame()
for code in flight:
    df = stock.get_market_cap("20200310", "20220228", code)['시가총액']
    
    df_flight_tt = pd.concat([df_flight_tt, df],axis=1)
    df_flight_tt.rename(columns={'시가총액':f'{code}시가총액'},inplace=True)
    time.sleep(1)

df_flight_tt['Sum'] = df_flight_tt.sum(axis=1)
df_flight_tt['index'] = (df_flight_tt['Sum']/df_flight_tt['Sum'].iloc[0]) *100
    
df_flight_tt
```

```python
index = pd.DataFrame({'makeup':df_makeup_tt['index'],
                      'travel':df_travel_tt['index'],
                      'flight':df_flight_tt['index']})
index.to_csv('makeup_travel_flight_index.csv')
```

```python
# 함수로 표현하기 : type에 종목 코드 리스트 입력
def stock_index(type):
    df_total = pd.DataFrame()
    for code in type:        
        df = stock.get_market_cap("20200310", "20220228", code)['시가총액']
        df_total = pd.concat([df_total, df],axis=1)
        df_total.rename(columns={'시가총액':f'{code}시가총액'},inplace=True)
        time.sleep(1)
    df_total['Sum'] = df_total.sum(axis=1)
    df_total['index'] = (df_total['Sum']/df_total['Sum'].iloc[0]) *100
    
    return df_total
```
